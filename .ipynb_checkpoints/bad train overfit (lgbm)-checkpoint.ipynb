```python
def convert_int(x):
    try:
        return int(x)
    except ValueError:
        return int(x[:-1])
```

```python
rf = pd.read_csv("MRegularSeasonCompactResults.csv")
seed_df = pd.read_csv("MNCAATourneySeeds.csv")
rf = rf.rename(columns={"WLoc":"ALoc","WTeamID":"ATeamID","WScore":"AScore","LTeamID":"BTeamID","LScore":"BScore"})
rf["y"] = 1
kf = rf
kf = kf.rename(columns={"BLoc":"ALoc","ATeamID":"BTeamID","AScore":"BScore","BTeamID":"ATeamID","BScore":"AScore"})
kf.y = 0
rf = pd.concat([rf,kf])
rf = rf.merge(seed_df, left_on=["ATeamID","Season"], right_on=["TeamID","Season"])
rf = rf.merge(seed_df, left_on=["BTeamID","Season"], right_on=["TeamID","Season"])
rf = rf.rename(columns={"Seed_x":"ASeed","Seed_y":"BSeed"}).drop(columns=["TeamID_x","TeamID_y"])
rf.loc[:,["AZone","ASeed"]] = list(rf.ASeed.apply(lambda x: [x[0],convert_int(x[1:])]))
rf.loc[:,["BZone","BSeed"]] = list(rf.BSeed.apply(lambda x: [x[0],convert_int(x[1:])]))
```
