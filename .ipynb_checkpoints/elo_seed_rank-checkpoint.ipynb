```python
data_dir = '../input/'
rs = pd.read_csv(data_dir + 'MRegularSeasonCompactResults.csv')
rs.head()
```

```python
def elo_pred(elo1, elo2):
    return(1. / (10. ** (-(elo1 - elo2) / 400.) + 1.))

def expected_margin(elo_diff):
    return((7.5 + 0.006 * elo_diff))

def elo_update(w_elo, l_elo, margin):
    elo_diff = w_elo - l_elo
    pred = elo_pred(w_elo, l_elo)
    mult = ((margin + 3.) ** 0.8) / expected_margin(elo_diff)
    update = K * mult * (1 - pred)
    return preds, update
```

```python
team_ids = set(rs.WTeamID).union(set(rs.LTeamID))
```

```python
rs["margin"] = rs["WScore"]-rs["LScore"]
```

```python
HOME_ADVANTAGE = 100
K = 22
```

```python
teamIDs = set(rs.WTeamID).union(set(rs.WTeamID)) #a set is a list in a dictionary with no duplicates and are ordered - union combines without duplicates
#https://www.w3schools.com/python/python_sets.asp
#teamIDs = set(list(rs.WTeamID) + list(rs.LTeamID))
#dict([(1,2),(3,2),(4,5)])
#rs.at[i, 'WTeamID']
```

```python
teamIDs = [(x,y) for y in teamIDs for x in range(1985,2021)]
```

```python
elo_dict = dict(zip(teamIDs,[1500 for x in range(len(teamIDs))])) #zip merges but can have set, list, or dict
```

```python
preds = []
```

```python
def elo(x):
    w = x['WTeamID']
    l = x['LTeamID']
    season = x["Season"]
    margin = x["margin"]
    wloc = x["WLoc"]
    
    w_ad, l_ad, = 0., 0.
    if wloc == "H":
        w_ad += HOME_ADVANTAGE
    elif wloc == "A":
        l_ad += HOME_ADVANTAGE
    
    
    pred, update = elo_update(elo_dict[(season,w)] + w_ad,
                              elo_dict[(season,l)] + l_ad, 
                              margin)
    elo_dict[(season,w)] += update
    elo_dict[(season,l)] -= update
    preds.append(pred)
    return [elo_dict[(season,w)], elo_dict[(season,l)]]
    
rs[["Welo","Lelo"]] = list(rs.apply(elo,axis=1))
```

```python
rs["preds"] = preds
```

```python
a = np.array(list(elo_dict))
b = np.array(list(elo_dict.values())).reshape(-1,1)
```

```python
elo_df = pd.DataFrame(np.concatenate((a,b),axis=1),columns=["Season","TeamID","Elo"])
```
